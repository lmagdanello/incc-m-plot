```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px


# Ler o arquivo CSV
df = pd.read_csv('./data/indices-incc-m.csv', delimiter=',', thousands='.', decimal=',')


# Extrair os meses
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']

# Converter os valores para números decimais
for mes in meses:
    df[mes] = df[mes].str.replace('%', '').str.replace(',', '.').astype(float)

# Transformar o dataframe no formato adequado para o Plotly
df_melted = df.melt(id_vars=['Anos'], value_vars=meses, var_name='Mês', value_name='Variação')

# Gerar o gráfico interativo
fig = px.line(df_melted, x='Mês', y='Variação', color='Anos',
              title='Variação do INCC por Ano - Mês',
              labels={'Mês': 'Mês', 'Variação': 'Variação', 'Anos': 'Ano'},
              hover_data=['Anos'])

# Configurar a legenda para aparecer sempre
fig.update_layout(legend=dict(yanchor="top", y=0.99, xanchor="left", x=0.01))

# Habilitar a seleção do ano através do botão de "toggle"
fig.update_traces(mode='lines+markers')

# Exibir o gráfico interativo
fig.show()
```
